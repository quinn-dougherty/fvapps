```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("df.csv")
df = df[df.lean_success]
```

```python
plt.boxplot(df[df.lean_success][["python_loops", "lean_loops"]], tick_labels=["Pytest Loops", "Lean Loops"])
plt.title("Number of loops it took to generate benchmark");
```

```python
df[df.lean_success][["python_loops", "lean_loops"]].value_counts()
```

```python
df.mean()
```

```python

max_loops = 5
table_data = []

for n in range(max_loops + 1):
    py_count = len(df[df['python_loops'] == n])
    lean_count = len(df[df['lean_loops'] == n])
    table_data.append([n, py_count, lean_count])

plt.figure(figsize=(8, 4))
plt.axis('off')
plt.table(cellText=table_data,
          colLabels=['Loops', 'Python Count', 'Lean Count'],
          loc='center',
          cellLoc='center')
plt.title('Distribution of Loop Counts')
plt.tight_layout()
plt.show()
```
